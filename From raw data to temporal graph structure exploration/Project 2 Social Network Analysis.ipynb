* We will now create a function that will recognize the authors and create the pairs of co-authors, while keeping a symmetry characteristic (x,y)=(y,x), which basically means that the order of the two authors in a pairs doesn't matter.

```python
def authors(df):
    for i in range(len(df)):
        df.Authors.iloc[i] = list(str(df.Authors.iloc[i]).split(";"))
    L=[]
    for j in list(df.Authors):
        list1 = list(combinations(j, 2))
        L.append(list1)
    new = []
    for k in range(len(L)):
        for l in range(len(L[k])):
              new.append(L[k][l])      
    df1 = pd.DataFrame(new)
    df1.columns = ['from', 'to']
    df1['weight'] = np.ones(len(new))
    df1['from'] = df1['from'].str.strip()
    df1['to'] = df1['to'].str.strip()
    final=(df1.groupby(df1[['from', 'to']].apply(lambda x: str(sorted(x)), axis=1)).agg({"from": "first", "to": "first", "weight": "sum"}).reset_index(drop=True))
    return final
```
